```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(9000)
```

```python
sr=100
ts=1/sr
t=np.arange(0,1,ts)
x=np.cos(2*np.pi*t)
plt.plot(t, x)
```

```python
def DFTi(x):
  N=len(x)
  X=[]
  for k in range(N):
    X_k=0
    for n in range(N):
      e=np.exp((2j*np.pi*k*n)/N)
      X_k+=x[n]/e
    X.append(X_k/(1j*k+0.00001))
  return np.array(X)
```

```python
k=np.arange(0, 100, 1)
```

```python
def IDFT(X):
  N=len(X)
  x=[]
  for n in range(N):
    x_n=0
    for k in range(N):
      e=np.exp((2j*np.pi*k*n)/N)
      x_n+=X[k]*e
    x.append(x_n/100)
  return np.array(x)
```

```python
X=DFTi(x)
V=IDFT(X)
plt.plot(t, V)
```

```python
PS=DFT(x)*np.conj(DFT(x))/len(DFT(x))
plt.plot(k, PS)
```

```python
def IcDFT(X):
  N=len(X)
  x=[]
  for n in range(N):
    x_n=0
    for k in range(N):
      e=np.exp((2j*np.pi*k*n)/N)
      if 6000>(X[k]*np.conj(X[k]))*N**(-1)>100:
        x_n+=X[k]*e
      else:
        x_n+=0
    x.append(x_n/100)
  return np.array(x)
```

```python
X=DFT(x)
V=IcDFT(X)
f_original = np.sin(2*np.pi*t)+5*np.sin(30*np.pi*t)
plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
plt.plot(t, f_original)

plt.subplot(1,4,2)
plt.plot(t, x)

plt.subplot(1,4,3)
plt.plot(t, V)

plt.subplot(1,4,4)
plt.plot(t, np.abs(np.mean(f_original) - V), color='r')
```
