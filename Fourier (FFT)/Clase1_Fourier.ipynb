```python
import numpy as np
from IPython.display import display, Audio
```

```python
# Sampling at 8 KHz
fs = 44100

# We'll make a pure tone at 220, and at 220 + 8000 = 8220
f0_original = 220
f0_aliased = f0_original + fs

# 2 seconds of audio should be plenty
duration = 3

# How many samples is 2 seconds?
N = int(duration * fs)

# Generate the sample positions
n = np.arange(N)

# Construct the first signal
x_original = np.cos(2 * np.pi * f0_original * n / fs)

# And the aliased signal
x_aliased = np.cos(10 * np.pi * f0_aliased * n / fs)

# Let's hear them both
print("fs = {} Hz, f0 = {} Hz".format(fs, f0_original))
display(Audio(data=x_original, rate=fs))

print("fs = {} Hz, f0 aliased = {} Hz".format(fs, f0_aliased))
display(Audio(data=x_aliased, rate=fs))
```

```python
duration = 1  # Duracin de cada nota en segundos
fs = 44100  # Frecuencia de muestreo
N = int(duration * fs)

# Generar las posiciones de muestra
n = np.arange(N)

do_octava = 16.35  # Frecuencia de do en la primera octava (C0)

# Calcular la frecuencia de do para cada octava y reproducir el sonido
for i in range(1, 10):  # Para cada octava (8 en total)
    f0 = do_octava * (2 ** (i - 1))  # Calcular la frecuencia de do para esta octava
    x_original = np.cos(2 * np.pi * f0 * n / fs)  # Generar la seal sinusoidal

    # Mostrar informacin y reproducir el sonido
    print("Octava {}, Frecuencia de do = {} Hz".format(i, f0))
    display(Audio(data=x_original, rate=fs))
```

```python
piano = np.zeros((8, 12))
frecuencia_actual = 16.35  # Frecuencia de la nota ms baja (C0)
nombres = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']

for i in range(8):
    for j in range(12):
        piano[i][j] = frecuencia_actual
        frecuencia_anterior = frecuencia_actual
        frecuencia_actual *= (2**(1/12))

        # Generar la seal sinusoidal para reproducir el sonido
        fs = 44100  # Frecuencia de muestreo
        duration = 1  # Duracin de cada nota en segundos
        N = int(duration * fs)
        n = np.arange(N)
        x_original = np.cos(2 * np.pi * frecuencia_actual * n / fs)

        # Mostrar informacin y reproducir el sonido
        octava_actual = i  # El ndice de la octava comienza en 0
        print("Nota: {}, Octava: {}, Frecuencia: {} Hz".format(nombres[j], octava_actual, frecuencia_anterior))
        display(Audio(data=x_original, rate=fs))

print(piano)
```

```python
la_mayor = (piano[3][9] + piano[4][1] + piano[4][4]) / 3

# Configurar parmetros de audio
fs = 44100  # Frecuencia de muestreo (generalmente se usa 44100 para audio)
duration = 4  # Duracin de cada nota en segundos
N = int(duration * fs)
n = np.arange(N)

# Generar seal de audio
x_original = (
    np.cos(2 * np.pi * piano[3][9] * n / fs)
    + np.cos(2 * np.pi * piano[4][1] * n / fs)
    + np.cos(2 * np.pi * piano[4][4] * n / fs)
) / 3

# Ajustar la amplitud para ser audible
x_amplified = x_original

# Mostrar informacin y reproducir audio
print("Acorde: {}, Octava: {}, Frecuencia: {} Hz".format("LaM", 4, la_mayor))
display(Audio(data=x_amplified, rate=fs))
```
