```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.integrate import odeint
from IPython.display import display
from ipywidgets import interactive,fixed
```

```python
alpha_0 = 3
# Definir las derivadas del sistema
def derivadas(X, alpha_0):
    x, y = X
    dotx = y
    doty = -x + x**3 - (alpha_0 + x)*y
    return np.array([dotx, doty])
```

```python
w_1 = -2
w_2 = 2
Y, X = np.mgrid[w_1:w_2:100j,w_1:w_2:100j]

U, V= derivadas([X,Y],alpha_0)

plt.streamplot(X, Y, U, V, density=[0.6,0.6], broken_streamlines=False)
plt.plot([-1],[0], marker = 'o', markersize=6, linestyle='', color='k')
plt.plot([0],[0], marker = 'o', markersize=6, linestyle='', color='k')
plt.plot([1],[0], marker = 'o', markersize=6, linestyle='', color='k')
plt.title("Varying density")
plt.grid()
plt.show()

p = interactive
```
