```python
def forward_diff(f,x,h):
    return (f(x+h)-f(x))/h

def backward_diff(f,x,h):
    return (f(x)-f(x-h))/h

def central_diff(f,x,h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
h = 0.001
funcion = np.sin
subconjunto = np.linspace(0, 1, int(h**(-1)))
print(forward_diff(funcion,subconjunto,h))
print(backward_diff(funcion,subconjunto,h))
print(central_diff(funcion,subconjunto,h))
```
